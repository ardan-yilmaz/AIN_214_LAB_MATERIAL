# 1. Line Plot 

* Used to display quantitative values over a continuous interval or time span. They are ideal for showing trends over time.

* **Advantages**
    - Clear display of trends over time.
    - Easy to compare multiple series.
* **Disadvantages**
    - Not ideal for discrete data categories.
    - Can be misleading if data is sparse. With limited data, as the line connects data points, it may imply a trend that does not exist.

```python
# Example of a Line Plot
years = [1950, 1970, 1990, 2010]
population = [2.519, 3.692, 5.263, 6.972]

plt.plot(years, population)
plt.title("World Population Over Years")
plt.xlabel("Year")
plt.ylabel("Population (in billions)")
plt.show()
```

# 2. Scatter Plot

- Used to display the relationship between two quantitative variables. They help in identifying the type of relationship, whether linear, quadratic, or no correlation, between the variables.

* **Advantages**
    - Effective for identifying and showing the relationship and distribution trends between two variables.
    - Useful in spotting outliers and anomalies in data.
* **Disadvantages**
    - Can become cluttered and less informative with very large datasets.
    -  Not ideal for displaying trends over time or with categorical data.


## Positive Correlation with Scatter Plots

```python
import matplotlib.pyplot as plt

# Sample data
heights = [150, 155, 160, 165, 170, 175, 180, 185]
weights = [50, 55, 60, 65, 70, 75, 80, 85]

plt.scatter(heights, weights)
plt.title("Height vs. Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```

## Negative (Inverse) Correlation (Vehicle Speed and Travel Time):


```python
import matplotlib.pyplot as plt

# Sample data
speeds = [20, 40, 60, 80, 100, 120]
travel_times = [90, 75, 60, 45, 35, 30]

plt.scatter(speeds, travel_times)
plt.title("Vehicle Speed vs. Travel Time")
plt.xlabel("Speed (km/h)")
plt.ylabel("Travel Time (min)")
plt.show()
```

## No Correlation (Random Scatter): 

```python
import matplotlib.pyplot as plt
import random

# Generating random sample data
shoe_sizes = [random.randint(35, 45) for _ in range(30)]
reading_skills = [random.randint(1, 100) for _ in range(30)]

plt.scatter(shoe_sizes, reading_skills)
plt.title("Shoe Size vs. Reading Skill")
plt.xlabel("Shoe Size (EU)")
plt.ylabel("Reading Skill (arbitrary units)")
plt.show()
```

## Outlier Detection

```python
import matplotlib.pyplot as plt
import random

# Generating sample data with one outlier
lengths = [random.uniform(14.8, 15.2) for _ in range(20)] + [16]  # Adding an outlier
widths = [random.uniform(7.8, 8.2) for _ in range(20)] + [9]     # Adding an outlier

plt.scatter(lengths, widths)
plt.title("Dimensions of Manufactured Items")
plt.xlabel("Length (cm)")
plt.ylabel("Width (cm)")
plt.show()
```

# 3. Histograms

* A graphical representation of the distribution of numerical data. It is an estimate of the probability distribution of a continuous variable.
* The primary use of a histogram is to visually show the distribution of a large set of data points. It helps in understanding where values are concentrated, their range, and if there are any gaps or unusual patterns in the data.
* Data is grouped into ranges (bins or buckets). Each bin represents a range of values, and the frequency of data points within each range is depicted by the height of the bar.
 Unlike bar charts which represent categorical data, histograms are used for continuous data, where the bins represent intervals
* **Advantages**
    - Easy to interpret and understand.
    - Can be used to identify the shape of the data distribution: normal distribution, skewed, bimodal, etc.
    - Can be used to identify outliers.
* **Disadvantages**
    - Loss of Data Specificity: As histograms group data into bins, they can obscure nuances in data distribution.
    - Sensitive to Bin Size: The choice of bin size can significantly affect the representation of data, potentially leading to misinterpretation.
    - No Multivariate Capability: Histograms are limited to showing the distribution of a single variable and cannot effectively represent relationships between multiple variables.

    


## Normal Distribution

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating sample data with a normal distribution
data = np.random.normal(loc=50, scale=10, size=1000)

plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## Skewed Distribution

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating sample data with a right-skewed distribution
data = np.random.exponential(scale=2, size=1000)

plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of Right-Skewed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## Bimodal Distribution

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating sample data with a bimodal distribution
data1 = np.random.normal(loc=20, scale=5, size=500)
data2 = np.random.normal(loc=60, scale=5, size=500)
data = np.concatenate((data1, data2))

plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of Bimodal Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

# 4. Bar Plots

* Used to compare discrete categories. Each bar represents a category.
* **Advantages**
    - Easily comparable 
    - Effective for both counts and proportions.Can be used to compare multiple categories.
* **Disadvantages**
    - Not ideal for continuous data: ineffective for showing trends or distributions over a continuous range.
    - Can become cluttered with too many categories.


```python
import matplotlib.pyplot as plt

# Example data
categories = ['Product A', 'Product B', 'Product C']
sales = [20, 35, 30]

plt.bar(categories, sales)
plt.title("Sales Comparison")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

# 5. Pie Charts

* Display parts-of-a-whole relationships, where each slice of the pie represents a proportion of the total.

* **Advantages**
    - Easy to understand and interpret.
    -  Highlight the size of categories relative to the whole.
* **Disadvantages**
    - Not ideal for comparing multiple categories.

```python
import matplotlib.pyplot as plt

# Example data
market_shares = {'Company A': 40, 'Company B': 30, 'Company C': 20, 'Company D': 10}
labels = market_shares.keys()
sizes = market_shares.values()

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Market Share Distribution")
plt.show()
```

# 6. Heat Maps

* Used to visualize complex data matrices. Each cell in the matrix is represented by a color, and the color intensity typically represents the magnitude or frequency of an event, highlighting patterns, variance, and anomalies in the data.

* Typically used for correlation matrix visualization, where each element in the matrix is the correlation coefficient between two variables, ranging from -1 to 1. 1 indicates a perfect positive correlation, -1 indicates a perfect negative correlation, and 0 indicates no correlation. The color intensity of each cell represents the magnitude of the correlation coefficient. 

* This visualization is particularly useful in fields like finance, research, and data science to understand how different variables affect each other, which can be crucial for predictive modeling and analysis.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generating a random dataset
data = np.random.rand(10, 10)

# Creating a correlation matrix
correlation_matrix = np.corrcoef(data)

# Creating a heatmap
sns.heatmap(correlation_matrix)
plt.title("Correlation Heatmap")
plt.show()
```
